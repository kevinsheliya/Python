```python
import logging

def basic_logger():
    logging.basicConfig(filename='app.log',level=logging.DEBUG)
    logging.debug("this is a debug message")
    logging.info("this is information message")
    logging.error("this is error message")
    logging.warning("this is warning message")
    logging.critical("this is a critical message")

basic_logger()
```

```python
def logging_with_diff_handling():
    logger=logging.getLogger("my_logger")
    logger.setLevel(logging.DEBUG)

    fil_handling=logging.FileHandler('app.log')
    console=logging.StreamHandler()

    fil_handling.setLevel(logging.DEBUG)
    console.setLevel(logging.DEBUG)

    formatted=logging.Formatter('%(asctime)s-%(name)s-%(levelname)s-%(message)s')
    fil_handling.setFormatter(formatted)
    console.setFormatter(formatted)

    logger.addHandler(fil_handling)
    logger.addHandler(console)

    logger.debug("this is a debug message")
    logger.info("this is information message")
    logger.error("this is error message")
    logger.warning("this is warning message")
    logger.critical("this is a critical message")


logging_with_diff_handling()

```

```python
import logging
from logging.handlers import RotatingFileHandler

def logger_with_handler():
    logger=logging.getLogger('routing_logger')
    logger.setLevel(logging.DEBUG)

    routing_handlers=RotatingFileHandler('rotating_app.log',maxBytes=2000,backupCount=5)
    formatted=logging.Formatter('%(asctime)s-%(name)s-%(levelname)s-%(message)s')
    routing_handlers.setFormatter(formatted)

    logger.addHandler(routing_handlers)

    for i in range(100):
        logger.debug("this is debug message number {}".format(i))

logger_with_handler()        
```

```python
def log_contextual():
    logger=logging.getLogger('context_logger')
    logger.setLevel(logging.DEBUG)

    file_hand=logging.FileHandler('context_logger.log')
    formatted=logging.Formatter('%(asctime)s-%(name)s-%(levelname)s-%(message)s')
    file_hand.setFormatter(formatted)

    logger.addHandler(file_hand)

    def message():
        logger.debug("This is a debug")
        logger.info("This is a info ")
        logger.error("This is a error")
        logger.critical("This is a critical")
        logger.warning("This is a warning")

    message()

log_contextual()       
```

```python
import logging
def context_log_arg(user_id,session_id):
    logger=logging.getLogger('context_log_arg')
    logger.setLevel(logging.DEBUG)

    fil_hand=logging.FileHandler('context_log_arg.log')
    formatted=logging.Formatter('%(asctime)s-%(name)s-%(levelname)s-%(message)s-%(funcName)s-%(lineno)d-UserId: %(user_id)s-SessionID:%(session_id)s')
    fil_hand.setFormatter(formatted)

    logger.addHandler(fil_hand)

    extra={'user_id' : user_id,'session_id':session_id}

    def test_log():
        logger.debug("This is a debug",extra=extra)
        logger.info("This is a info ",extra=extra)
        logger.error("This is a error",extra=extra)
        logger.critical("This is a critical",extra=extra)
        logger.warning("This is a warning",extra=extra)

    test_log()


context_log_arg('user123','session123')    




```

```python
import logging.config

def configure_logging_with_dict():
    log_config = {
        'version': 1,
        'formatters': {
            'default': {
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
            },
            'detailed': {
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s - %(funcName)s - %(lineno)d'
            }
        },
        'handlers': {
            'file': {
                'class': 'logging.FileHandler',
                'filename': 'dict_config_app.log',
                'formatter': 'detailed',
                'level': 'DEBUG'
            },
            'console': {
                'class': 'logging.StreamHandler',
                'formatter': 'default',
                'level': 'DEBUG'
            }
        },
        'root': {
            'handlers': ['file', 'console'],
            'level': 'DEBUG'
        }
    }
    logging.config.dictConfig(log_config)
    logger = logging.getLogger('')
    logger.debug('This is a debug message')
    logger.info('This is an info message')
    logger.warning('This is a warning message')
    logger.error('This is an error message')
    logger.critical('This is a critical message')

# Test the function
configure_logging_with_dict()
```
