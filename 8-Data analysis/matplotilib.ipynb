```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# line designing 
plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth='2',markersize='9')
plt.grid(True)
```

```python
# multipal plot
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,7))

plt.subplot(2,2,1)
plt.plot(x,y1,color='red',marker='o')
plt.title("Plt 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue',marker='o')
plt.title("plt 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='green',marker='o')
plt.title("plt 3")

plt.subplot(2,2,4)
plt.plot(y2,y1,color='yellow',marker='o')
plt.title('plt 4')


```

```python
## bar plot
catagories=['A','B','C','D','E']
values=[5,3,4,7,8]

#create bar plot
plt.bar(catagories,values,color='darkblue')
plt.xlabel("catagories")
plt.ylabel("values")
plt.title("Bar plot")
plt.show()
#plt.grid(True)

```

```python
## Histogrmas plot
data=[1,2,2,3,3,3,4,4,5,5,5,5,6]

plt.hist(data,bins=6,color='red',edgecolor='black')
```

```python
## scatter plot
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.scatter(x,y,color='blue',marker='x')
```

```python
##pip chart

labels=['A','B','C','D']
values=[20,35,15,30]
colors=['gold','lightskyblue','yellowgreen','lightcoral']
explode=(0.1,0,0,0.2) # move slice

plt.pie(values,explode=explode,labels=labels,colors=colors,autopct="%1.0f%%",shadow=True)
```

```python
# saleing the data
import pandas as pd
sales_data_df=pd.read_csv("sales_data.csv")
sales_data_df.head(5)

```

```python
# plot sale 
sales_data_df.info()
```

```python
T_sales_product=sales_data_df.groupby('Product Category')['Units Sold'].sum().reset_index()
print(T_sales_product)
```

```python
T_sales_product.plot(kind='bar',color='teal')
```

```python
## plot sales over time
sales_trend=sales_data_df.groupby('Date')['Units Sold'].sum().reset_index()
print(sales_trend)
```

```python
plt.plot(sales_trend['Date'],sales_trend['Units Sold'])
```
