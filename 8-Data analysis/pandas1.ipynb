```python
# miising value fill with mean Colums
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df
```

```python
## REname colums
df=df.rename(columns={'Date':'Sale_data'})
df.head(5)
```

```python
#chage datatype
df['New_value']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['Value_new']=df['Value'].apply(lambda x:x*2)
df.head()
```

```python
# Data Aggregating and grouping
df.head()
```

```python
group_mean=df.groupby('Product')['Value'].mean()
print(group_mean)
```

```python
group_sum=df.groupby(['Product','Region'])['Value'].sum()
print(group_sum)
```

```python
## Multipal aggregate function

group_agg=df.groupby('Region')['Value'].agg(['mean','sum','count','std'])
print(group_agg)
```

```python
## Merging and joining dataframes

df1=pd.DataFrame({'key':['A','B','C'],'Value1':['1','2','3'],'val1':['4','5','6']})
df2=pd.DataFrame({'key':['A','B','D'],'Value2':['7','8','9'],'val2':['10','11','12']})
```

```python
df1
```

```python
df2
```

```python
## Merging on  KEY columns 

pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
