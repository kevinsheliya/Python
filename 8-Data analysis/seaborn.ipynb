```python
import seaborn as sns
```

```python
# Basic plotting 
tips=sns.load_dataset('tips')
tips
```

```python
#scatter plot 

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plot pf total bill vs tip ")
plt.show()
```

```python
#line plot 
sns.lineplot(x='size',y='tip',data=tips)
plt.title("line plot of tip by  size")
plt.show()
```

```python
# categorical plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("categorical plot days vs total bill")
plt.show()
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## histogram
sns.histplot(tips['total_bill'],bins=12,kde=True)
```

```python
# kde plot 
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
# pairple plot
sns.pairplot(tips)
```

```python
# heat map
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
